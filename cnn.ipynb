```python
import emoji
import re
def preprocessing(text):
# Uniformity of case  
  text = text.lower()

# Remove punctuation
  exclude = string.punctuation
  text = text.translate(str.maketrans('', '', exclude))

# Remove numbers
  text = ''.join([i for i in text if not i.isdigit()])

# Remove user
  text = (text.replace('user',''))

# Remove emoji
  text = emoji.demojize(text)
  text = re.sub(':\S+?:', ' ', text)

# Remove_stopwords
# wnl = WordNetLemmatizer() 
  s = [w for w in text.split(' ') if w not in stop]
  return ' '.join(s)
```

## Tokenize and encode
We combine the train, validation, and test dataset in order to use the same word-to-index.

```python
import spacy
nlp = spacy.blank("en")
def tokenize(texts, max_len):
  max_len = max_len
  tokenized_texts = []
  word2idx = {}

  # Add <pad> and <unk> tokens to the vocabulary
  word2idx['<pad>'] = 0
  word2idx['<unk>'] = 1

  # Building our vocab from the corpus starting from index 2
  idx = 2
  for sent in texts:
    tokenized_sent = nlp(sent)
    # Add `tokenized_sent` to `tokenized_texts`
    tokenized_texts.append(tokenized_sent)
    # Add new token to `word2idx`
    for token in tokenized_sent:
      # string any token objects are different things, be careful.
      if token.text not in word2idx:
        word2idx[token.text] = idx
        idx += 1

        # Update `max_len`
    max_len = max(max_len, len(tokenized_sent))

  return tokenized_texts, word2idx, max_len
```

```python
def encode(tokenized_texts, word2idx, max_len):
    input_ids = []
    for tokenized_sent in tokenized_texts:
        # Pad sentences to max_len
        tokenized_padded_sent = list(tokenized_sent) + ['<pad>'] * (max_len - len(tokenized_sent))
 
        # Encode tokens to input_ids
        input_id = [word2idx.get(str(token)) for token in tokenized_padded_sent]
        input_ids.append(input_id)
    
    return np.array(input_ids)
```

#CNN
Here we build the CNN model, and use the tuner(random search, trials = 15) to find the best parameters

```python
from tensorflow import keras
from tensorflow.keras import layers
from kerastuner.tuners import RandomSearch
import keras_tuner as kt
from keras.layers import BatchNormalization
from sklearn.metrics import f1_score
#import tensorflow_lattice as tfl
```

```python
def build_model(hp):
    model = keras.Sequential([
        layers.Input(shape=(50,)),
        layers.Embedding(20000,50),
        layers.Conv1D(filters = hp.Int('filters',
                                        min_value=64,
                                        max_value=256,
                                        step = 32),
               kernel_size = hp.Int('kernel_size',
                                        min_value=2,
                                        max_value=16,
                                        step = 2),
               strides=hp.Int('strides',
                                        min_value=2,
                                        max_value=6,
                                        step = 2),
               padding = "same",
               activation = 'relu'),
        layers.MaxPool1D(),
        layers.Flatten(),
        layers.Dropout(rate=hp.Float('rate',min_value=0.1,max_value=0.5,step=0.2)),
        layers.Dense(1, activation = 'sigmoid')  
    ]) 
    model.compile(
    optimizer=keras.optimizers.Adam(
        hp.Choice('learning_rate',
                  values=[1e-2, 1e-3, 1e-4])),
    loss='binary_crossentropy',
    metrics=['accuracy'])
    return model
```

```python
tuner = RandomSearch(
    build_model,
    objective='val_accuracy',
    max_trials=15
  )
tuner.search_space_summary()
```

```python
anger_joy_trainval_label = np.asarray(anger_joy_trainval_label).astype(np.int64)
```

```python
tuner.search(input_ids_anger_joy_trainval, anger_joy_trainval_label,
             epochs=10,
             validation_split=0.2
             )

model_best = tuner.get_best_models(num_models=2)[0]
```

```python
tuner.results_summary(num_trials=15)
```

```python
hps = tuner.get_best_hyperparameters(num_trials=15)
for i in range(15):
  print('-'*50)
  print("This is the round :", i+1)
  print("number of filters is :",hps[i].get('filters'))
  print("kernal size is :",hps[i].get('kernel_size'))
  print("strides is :",hps[i].get('strides'))
  print("dropout is :",hps[i].get('rate'))
  print("learning rate is :",hps[i].get('learning_rate'))
  model_test = tuner.hypermodel.build(hps[i])
  history = model_test.fit(input_ids_anger_joy_trainval, anger_joy_trainval_label, epochs=10, validation_split=0.2)
  val_acc_per_epoch = history.history['val_accuracy']
  best_epoch = val_acc_per_epoch.index(max(val_acc_per_epoch)) + 1
  hypermodel = tuner.hypermodel.build(hps[i])
  hypermodel.fit(input_ids_anger_joy_trainval, anger_joy_trainval_label, epochs=best_epoch, validation_split=0.2)
  print("f1-macro is :", f1_score(anger_joy_trainval_label, hypermodel.predict(input_ids_anger_joy_trainval).round(), average='macro'))
```
